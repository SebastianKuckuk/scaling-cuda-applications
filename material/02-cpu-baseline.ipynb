[02-cpu-baseline/stencil-2d.cpp](../src/02-cpu-baseline/stencil-2d.cpp) implements the algorithm as a *serial* *CPU-only* application.

```python
!../build/02-cpu-baseline
```

At runtime, the program:

* allocates two 2D arrays, `u` and `uNew`,
* initializes the temperature field with a heat pulse, and sets the boundaries to zero,
* iterates through time steps and for each step
  * computes a new temperatures using a 5-point stencil in each cell of the grid,
  * swaps `u` and `uNew`,
* optionally writes temperature fields to disk for visualization.

The application accepts several arguments to configure the simulation domain and runtime behavior:

* `globalNumCellsX`, the number of grid cells in the x-direction including boundary cells
* `globalNumCellsY`, the number of grid cells in the y-direction including boundary cells
* `nWarmUp`, the number of **warm-up iterations** executed before timing begins - used to eliminate startup effects
* `numItTimed`, the number of **timed iterations** (main simulation loop) executed for measurement
* `outputInterval` which controls how often output is written - a value of `0` disables output completely; otherwise, file output is generated every `outputInterval` iterations
